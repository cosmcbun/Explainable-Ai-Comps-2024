# Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_excel("Questions asked in both surveys.xlsx")
df.head()
```

# Recode

```python
def code_experence_ml(experience:str):
	if "Completed CS 320" in experience:
		return "5:Completed ML Course"
	elif "Currently taking CS" in experience:
		return "4:Enrolled In ML Course"
	elif "Extracurricular experience" in experience:
		return "3:Extracurricular Experience in ML"
	elif "Experience with generative AI" in experience:
		return "2:Dabbled in ML"
	elif "dabbeled" in experience:
		return "2:Dabbled in ML"
	elif "dasbbled" in experience:
		return "2:Dabbled in ML"
	elif "high school" in experience:
		return "2:Dabbled in ML"
	elif "No experience" in experience:
		return "1:No ML Experience"
	else:
		return 100

df["Coded Experience with ML"] = df["Experience with ML"].apply(code_experence_ml)
df[["Experience with ML", "Coded Experience with ML"]]
```

```python
def code_experence_ml_three(experience:str):
	if "5" in experience:
		return "Formal ML Experience"
	elif "4" in experience:
		return "Formal ML Experience"
	elif "3" in experience:
		return "Informal ML Experience"
	elif "2" in experience:
		return "Informal ML Experience"
	else:
		return "No ML Experience"

df["Simple Coded Experience with ML"] = df["Coded Experience with ML"].apply(code_experence_ml_three)
df[["Experience with ML", "Simple Coded Experience with ML"]]
```

```python
def get_rankings(technique:str):
	new = df[[f"Favorite Explanation [{technique}]", "Which survey are you administering?", "Coded Experience with ML", "Simple Coded Experience with ML"]]
	new = new.rename(columns={f"Favorite Explanation [{technique}]": "Ranking"})
	new["Ranking"] = pd.to_numeric(new["Ranking"].str[:-2], errors='coerce')
	new["Ranking"] = abs(new["Ranking"] - 4)
	new["Technique"] = technique
	return new

randking_df = pd.concat([get_rankings(i) for i in ["Anchor", "Shapley", "LIME"]])
# save randking_df as xlsx
randking_df.to_excel("rankings.xlsx")
randking_df
```

# Plots

## Favorite Explanations

```python
num_favorite_anchor, num_favorite_lime, num_favorite_shap = [df[f"Favorite Explanation [{i}]"].value_counts()["1st"]for i in ["Anchor", "LIME", "Shapley"]]
title = "Favorite Explanation Method"

plt.pie(
	[num_favorite_anchor, num_favorite_shap, num_favorite_lime],
	labels=["Anchor", "Shapley", "LIME"],
	autopct="%1.1f%%")
plt.title(title)
plt.savefig(f"{title}.png", bbox_inches="tight")
plt.show()
```

```python
num__least_favorite_anchor, num_least_favorite_lime, num_least_favorite_shap = [df[f"Favorite Explanation [{i}]"].value_counts()["3rd"] for i in ["Anchor", "LIME", "Shapley"]]
title = "Least Favorite Explanation Method"

plt.pie(
	[
		num__least_favorite_anchor,
		num_least_favorite_shap,
		num_least_favorite_lime
	],
	labels=["Anchor", "Shapley", "LIME"],
	autopct="%1.1f%%")
plt.title(title)
plt.savefig(f"{title}.png", bbox_inches="tight")
plt.show()
```

```python
title = "Do the Different Surveys Affect Rankings of XAI Techniques"

g = sns.catplot(
    data=randking_df,
	kind="bar",
    x="Technique",
	y="Ranking",
	hue="Which survey are you administering?",
    errorbar="sd",
	height=6
).set_axis_labels(
	"XAI Technique",
	"Ranking (3 is favorite, 1 is least favorite)"
)

plt.title(title+"?")
plt.savefig(f"{title}.png", bbox_inches="tight")
plt.show()
```

```python
title = "Do Participant Experence Affect Rankings of XAI Techniques"

g = sns.catplot(
    data=randking_df,
	kind="bar",
    x="Technique",
	y="Ranking",
	hue="Simple Coded Experience with ML",
    errorbar="sd",
	height=6
).set_axis_labels(
	"XAI Technique",
	"Ranking (3 is favorite, 1 is least favorite)"
)
plt.title(title+"?")
# when saving the plot, the title is cut off
plt.savefig(f"{title}.png", bbox_inches="tight")
plt.show()
```

## Trust in Explanations

```python

def get_trust_chart(condition, technique):
	counter = df\
		[df["Which survey are you administering?"] == condition]\
		[f"How much do you trust each of these techniques to explain a model's thought process in new examples? [{technique}]"]\
		.value_counts()
	plt.pie(
		counter.to_list(),
		labels=counter.index.to_list(),
		autopct="%1.1f%%"
	)
	plt.title(f"Trust in {technique}")
	return plt

def get_trust_charts(condition):
	title = f"Trust from {condition} particpants"

	plt.subplot(1, 3, 1)
	get_trust_chart(condition, "Anchor")

	plt.subplot(1, 3, 2)
	get_trust_chart(condition, "Shapley")

	plt.subplot(1, 3, 3)
	get_trust_chart(condition, "LIME")


	plt.gcf().set_size_inches(12, 4)
	plt.suptitle(title)
	plt.savefig(f"{title}.png", bbox_inches="tight")
	plt.show()

get_trust_charts("Mooc")
get_trust_charts("Resnet")
```
